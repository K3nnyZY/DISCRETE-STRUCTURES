#### Se Lanzan $n$ dados $T=10^i$ veces. $i=1, 2, 3$. Muestre los 3 gráficos de barras correspondientes.


```python
n = 2  # numero de dados
T_values = [10**i for i in range(1, 4)] # veces de lanzamientos

for T in T_values:
    fa = np.zeros(5*n+1)
    for i in range(T):
        fa[np.sum(np.random.randint(1, 7, n)) - n] += 1
    
    # Normaliza las frecuencias para convertirlas en probabilidades
    fa /= T

    # Crea un gráfico de barras con las probabilidades
    plt.bar(range(n, 6*n+1), fa)
    plt.title(f'Lanzamiento de {n} dados {T} veces')
    plt.xlabel('Suma de los dados')
    plt.ylabel('Probabilidad')
    plt.grid(axis='y')
    plt.show()
```

### Graficas en 2D
pasos:

1. Dar valores a la variable independiente (ejex)
2. evaluar ejey = f(ejex)
3. graficar

```python
# Paso 1: Dar valores a la variable independiente (eje x)
x = np.linspace(-10, 10, 400)

# Paso 2: Evaluar eje y = f(eje x)
# Usamos la función f(x) = x^2 como ejemplo
y = x**2

# Paso 3: Graficar
plt.plot(x, y)

# Añadimos etiquetas y título para mayor claridad
plt.xlabel('eje x')
plt.ylabel('eje y')
plt.title('Gráfico de $f(x) = x^2$')
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

# Mostramos la gráfica
plt.grid(True)
plt.show()
```

### coordenadas polares
Convierte la ecuación cartesiana
$$x^2 + y^2 = 4$$
a coordenada polares y graficala

```python
# despejando r tenemos que r = 2
theta = np.linspace(0, 2*np.pi, 300)
r = 2

# Convertimos las coordenadas polares a coordenadas cartesianas
x = r * np.cos(theta)
y = r * np.sin(theta)

# Graficamos el círculo
plt.plot(x, y)

# Añadimos etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de $r=2$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Mostramos el gráfico
plt.show()
```

## Grafica en 3D
pasos: 
1. Dar valores a la variable independiente (ejex,ejey)
2. Hacer el producto cartesiano (ejex,ejey,ejez),  es decir la rejilla
3. evaluar ejez = f(ejex,ejey)
4. graficar

### Coordenadas rectangulares
Grafica la funcion:
$$z = -x^2 - y^2 + 4$$
sobre el dominio $[-2,2]$, para $x$ y $[-2,2]$ para $y$.

```python
# Paso 1: Dar valores a las variables independientes (eje x, eje y)
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

# Paso 2: Hacer el producto cartesiano (eje x, eje y, eje z), es decir, la rejilla
X, Y = np.meshgrid(x, y)

# Paso 3: Evaluar eje z = f(eje x, eje y)
# Usamos la función f(x, y) = sin(sqrt(x^2 + y^2)) como ejemplo
Z = -X**2 - Y**2 +4

# Paso 4: Graficar
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Añadimos etiquetas y título para mayor claridad
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Gráfico 3D de $f(x, y) = -x^2 - y^2 +4$')

# Mostramos la gráfica
plt.show()
```

### Coordenadas cilindricas

Grafica la funcion:
$$z = r^2 cos(3\theta)e^{-r}$$
sobre el dominio $r \in [0,2]$ y $\theta \in [0,2\pi]$ 

```python
# Paso 1: Definir el dominio
r = np.linspace(0, 2, 100)
theta = np.linspace(-2, 2*np.pi, 100)

# Paso 2: Crear la malla, producto cartesiano
R, Theta = np.meshgrid(r, theta)

# Paso 3: Evaluar la función
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

z = r**2 * np.cos(3*theta) * np.exp(-r)

# Paso 4: Graficar
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Añadimos etiquetas y título para mayor claridad
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# Mostramos la gráfica
plt.show()
```
